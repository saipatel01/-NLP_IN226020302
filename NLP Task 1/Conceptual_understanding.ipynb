Task 1: Conceptual Understanding

1. Difference between "Love" and "love" in NLP

   In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.

   "Love" may appear at the beginning of a sentence or as a proper noun
   "love" is the standard lowercase form

   Without lowercasing, the model treats them as separate words → increases vocabulary size and reduces learning efficiency.



2. What happens if stopwords are not removed?

   If stopwords are not removed:

   The dataset becomes noisy and large
   
   Common words like "the", "is", "and" dominate frequency counts
   
   Model performance may degrade because meaningful words get less importance

3. When removing stopwords is harmful 

   I. Sentiment Analysis

      Example: "I am not happy"

      Removing "not" → "I happy" → wrong sentiment

   II. Question Answering / Search Systems

       Example: "What is the best laptop?"
     
       Removing stopwords may distort query meaning

4. Difference between Stemming and Lemmatization

Feature 	 |    Stemming        	       |        Lemmatization

Method	     |  Rule-based cutting	        |      Dictionary-based

Output	     |  May not be real word        |     Always valid word

Example 	 |  "running" → "run" / "runn"   |     "running" → "run"

Accuracy	 |   Lower	                     |     Higher
