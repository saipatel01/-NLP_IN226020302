```python
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Remove URLs & emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)

    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()

    # Remove repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters / emojis
    text = re.sub(r'[^\w\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short tokens (≤2) except meaningful ones
    exceptions = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence 
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)

    result = {
        "original": sentence,
        "tokens": tokens,
        "clean": clean
    }
    results.append(result)

for r in results:
    print("\nOriginal:", r["original"])
    print("Tokens:", r["tokens"])
    print("Clean Sentence:", r["clean"])
```

```python
import numpy as np

analytics = []

for r in results:
    tokens = r["tokens"]
    
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(t) for t in tokens]) if tokens else 0

    analytics.append({
        "sentence": r["original"],
        "total_tokens": total_tokens,
        "unique_tokens": unique_tokens,
        "avg_token_length": avg_length
    })

for a in analytics:
    print("\nSentence:", a["sentence"])
    print("Total Tokens:", a["total_tokens"])
    print("Unique Tokens:", a["unique_tokens"])
    print("Avg Length:", round(a["avg_token_length"], 2))
```

```python
from collections import Counter

all_tokens = []

for r in results:
    all_tokens.extend(r["tokens"])

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[:-6:-1])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("\nInput:", case)
    print("Tokens:", tokens)
    print("Clean:", clean)
```
